## Mixtures of Guassians

Sometimes you have a distribution of values that is not unimodal, but is multi-modal.  In this case we can use a mixture of Guassians to model this distribution.  Her's how.

For example lets assume that we have a population that we would like to model as being comprized of 2 subpopulations.  Thus we introduce a new paramater, $\pi \in (0,1)$, which will represent the proportion of our entire population that is in the first subpopulation, and thus $1-\pi$ will represent the proportion that is in the second sub-population.  (In general we will need to look at a partition $\pi_i$ of $(0,1)$ with $n$ partitions for each mode in the data that we want to try to capture).  Then we make the assumption that the two subpopulations come from normal distributions $\mathcal{N}(\mu_1, {\sigma_1}^2)$ and $\mathcal{N}(\mu_2, {\sigma_2}^2)$.  Then the PDF of the new distribution is given by 
$$ f(x) = \pi\frac{1}{\sigma_1\sqrt{2\pi}}\exp(-\frac{(x-\mu_1)^2}{2{\sigma_1}^2}) + (1-\pi)\frac{1}{\sigma_2\sqrt{2\pi}}\exp(-\frac{(x-\mu_2)^2}{2{\sigma_1}^2})$$

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math as math
```

```python
from matplotlib.widgets import Slider, Button
```

```python
mu_1 = 0
variance_1 = 1
sigma_1 = math.sqrt(variance_1)

mu_2 = 3
variance_2 = 2
sigma_2=math.sqrt(variance_2)

pi = 0.3

x = np.linspace(mu_1 - 3*sigma_1,mu_2 + 3*sigma_2,100)
y_1 = stats.norm.pdf(x, mu_1, sigma_1)
y_2 = stats.norm.pdf(x, mu_2, sigma_2)

mixture = pi*(y_1) + (1-pi)*(y_2)

plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,mixture)
plt.show
```
