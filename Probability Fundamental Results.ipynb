# Strong and weak law of large numbers

Let $X_1,\dots,Xn$ be i.i.d. random variables, with $\mu=\mathbb{E}[X]$ and $\sigma^2=\text{var}[X]$.

Laws (weak and strong) of large numbers (LLN):

$$
\bar{X}_n:=\frac{1}{n}\sum_{i=1}^n X_i \substack{\text{P, a.s.}\\{\longrightarrow}\\{n\to\infty}} \mu
$$    	 

where the convergence is in probability (as denoted by P on the convergence arrow) and almost surely (as denoted by a.s. on the arrow) for the weak and strong laws respectively.  (See modes of convergence below.)

### Central limit theorem (CLT):

$$
\sqrt{n}\frac{\bar{X}_n−\mu}{\sigma} \substack{\text{(d)}\\{\longrightarrow}\\{n\to\infty}}	\mathcal{N}(0,1) 	 	 
$$

or equivalently,

$$
\sqrt{n}(\bar{X}_n−\mu) \substack{\text{(d)}\\{\longrightarrow}\\{n\to\infty}}	\mathcal{N}(0,\sigma^2)  	 	 
$$

where the convergence is in distribution, as denoted by (d) on top of the convergence arrow. 

We will revisit the different modes of convergence near the end of this lecture.

Note : In 6.431x: Probability–the Science of Uncertainty and Data, we used yet another equivalent formulation of the CLT:
 	
$$
\frac{S_n−n\mu}{\sqrt{n}\sigma}\substack{\text{(d)}\\{\longrightarrow}\\{n\to\infty}}	\mathcal{N}(0,1)	 	 
$$

where $S_n=\sum_{i=1}^n X_i$ is the sum (not the average) of $X_i$. 

### Hoeffding's Inequality:

Let $X_1,\dots,X_n$ be i.i.d random variables sucht that $\mu=\mathbb{E}[X]$ and such that $X\in[a,b]$ almost surely (where $a<b \in \mathbb{R}$.  Then

$$
\mathbb{P}[|\bar{X}_n -\mu| \ge \epsilon] \le 2\exp{\left( -\frac{2n\epsilon^2}{(b-a)^2}\right)}, \forall \epsilon \gt 0
$$

### Convergence

There are three types of probabilistic or stochastic convergence.  The set up is this: If $T_1,\dots,T_n$ is a sequence of random variables and $T$ is another random variable (which might be deterministic) then we can define three types of convergence:

#### Convergence Almost Surely

We say the sequence  $T_1,\dots,T_n$ converges almost surely to $T$ iff

$$
T_n \substack{\text{a.s.}\\\longrightarrow\\n\to\infty} T \iff \mathbb{P}\left[ \lbrace \forall \omega : T_n(\omega) \substack{\longrightarrow\\n\to\infty}T(\omega)\rbrace\right] = 1
$$

This is telling us that no matter what element $T_n(\omega)$ we chose in our sample space, as $n\to\infty$ then $T_n(\omega) \to T(\omega)$

#### Convergence in probability:

$$
T_n \substack{\mathbb{P}\\\longrightarrow\\n\to\infty} \iff \mathbb{P}\left[ |T_n - T|\ge \epsilon\right]\substack{\longrightarrow\\n\to\infty} 0, \forall \epsilon \gt 0
$$

This is telling us that for any $\epsilon$ the probability that $T_n$ is more than $\epsilon$ away from $T$ tends to 0. 

#### Convergence in distribution:

$$
T_n \substack{\text{(d)}\\\longrightarrow\\n\to\infty} \iff \mathbb{E}\left[ f(T_n) \right]\substack{\longrightarrow\\n\to\infty} \mathbb{E}\left[f(T)\right]
$$

for all continuous bounded functions $f$.

The three have a hierarchy,  convergence almost surely $\implies$ convergence in probability $\implies$ convergence in distribution.  Even though convergence in distribution is the weakest of the three, it is probably the most used type of convergence.

## Properties

Addition, multiplication and division preserve convergence almost surely $\text{a.s.}$ and convergence in probability $\mathbb{P}$.  In particular if we have

$$
T_n \substack{\text{a.s.}/\mathbb{P}\\\longrightarrow\\n\to\infty} T \text{ and } U_n \substack{\text{a.s.}/ \mathbb{P}\\\longrightarrow\\n\to\infty} U 
$$

Then

$$
T_n + U_n \substack{\text{a.s.}/\mathbb{P}\\\longrightarrow\\n\to\infty} T + U
$$

$$
T_n U_n \substack{\text{a.s.}/\mathbb{P}\\\longrightarrow\\n\to\infty} T U
$$

and if in addition $U\ne0$ then

$$
\frac{T_n}{U_n} \substack{\text{a.s.}/\mathbb{P}\\\longrightarrow\\n\to\infty} \frac{T}{U}
$$

Note that in general these do not apply to convergence in distribution $(d)$.  In the case of convergence in distribution we have the following very useful result.



#### Slutsky's Theorem

Let $(X_n), (Y_n)$ be two sequences of random variables such that:

$$
T_n \substack{\text{(d)}\\\longrightarrow\\n\to\infty} T \text{ and } U_n \substack{\text{(d)}\\\longrightarrow\\n\to\infty} u
$$

Where $T$ is a random variable and $u$ is a real number. That is $U_n$ has a deterministic limit:  $\mathbb{P}(U = 1) = 1$

Then:

$$
T_n + U_n \substack{\text{(d)}\\\longrightarrow\\n\to\infty} T + u
$$

$$
T_n U_n \substack{\text{(d)}\\\longrightarrow\\n\to\infty} T u
$$

and if in addition $U\ne0$ then

$$
\frac{T_n}{U_n} \substack{\text{(d)}\\\longrightarrow\\n\to\infty} \frac{T}{u}
$$

### Continuous mapping theorem

If $f$ is a continuous function then

$$
T_n \substack{\text{a.s.}/\mathbb{P}/\text{(d)}\\\longrightarrow\\n\to\infty} T \implies f(T_n) \substack{\text{a.s.}/\mathbb{P}/\text{(d)}\\\longrightarrow\\n\to\infty} f(T) 
$$
